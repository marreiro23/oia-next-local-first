```python
# Python script for interacting with local Ollama RAG

import requests
import json

# Base URL for Ollama API
base_url = "http://localhost:11434/api"

def query_ollama(prompt, model="llama3", context=None):
    """
    Query the local Ollama RAG system
    """
    url = f"{base_url}/generate"
    
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }
    
    # Add context if provided (for RAG)
    if context:
        payload["context"] = context
    
    response = requests.post(url, data=json.dumps(payload))
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# Example usage
if __name__ == "__main__":
    # Simple query
    result = query_ollama("Explain quantum computing in simple terms")
    if result:
        print("Response:", result['response'])
    
    # RAG query with context
    rag_context = ["Quantum computing uses quantum bits (qubits) that can exist in multiple states simultaneously.",
                   "This allows quantum computers to solve certain problems much faster than classical computers."]
    result = query_ollama("Explain quantum computing based on the provided context", 
                         model="llama3-rag", 
                         context=rag_context)
    if result:
        print("RAG Response:", result['response'])
```
